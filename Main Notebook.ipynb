**1.2. Dataset Description**

The dataset provided for customer segmentation analysis consists of the following variables:

1. ID (numerical): Unique identifier for each customer.

2. Sex (categorical): Biological sex (gender) of the customer. Two options available:
0: Male
1: Female

3. Marital Status (categorical): Marital status of the customer. Two options available:
0: Single
1: Non-single (divorced/separated/married/widowed)

4. Age (numerical): Age of the customer in years, calculated as the current year minus the year of birth.

5. Education (categorical): Level of education of the customer. Four options available:
0: Other/Unknown
1: High School
2: University
3: Graduate School

6. Income (numerical): Self-reported annual income of the customer in US dollars.

7. Occupation (categorical): Category of occupation of the customer. Three options available:
0: Unemployed/Unskilled
1: Skilled Employee/Official
2: Management/Self-employed/Highly Qualified Employee/Officer

8. Settlement Size (categorical): The size of the city that the customer lives in. Three options available:
0: Small City
1: Mid-sized City
2: Big City
The dataset consists of 2000 customer entries, and each customer has corresponding values for the aforementioned variables. The ranges for numerical variables are specified, along with the possible values for categorical variables.

To perform customer segmentation analysis on the provided dataset, we will utilize two clustering techniques: k-Means and Hierarchical clustering.

## 4. Customer Segmentation Analysis

**4.1 k-Means method**

**Import kmeans and standardise data**

```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

# Drop the ID column
df = df.drop('ID', axis = 1)
```

**Using the Elbow method to find the optimal number of clusters {-}**

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Elbow method to find the optimal number of clusters
inertias = []  # empty list

for i in range(1, 11):
    km = KMeans(n_clusters=i, init='k-means++', n_init=10, max_iter=300, random_state=0)
    km.fit(df)
    inertias.append(km.inertia_)
    print(km.inertia_)

plt.plot(range(1, 11), inertias, marker='o', color='orange')
plt.xlabel('The number of clusters')
plt.ylabel('Cluster inertia (within-cluster SSE)')
plt.xticks(range(1, 11))
plt.tight_layout()

# Set the title
plt.title('Elbow Method: Cluster Inertia vs. Number of Clusters')

plt.show()
```

**The elbow plots show that the optimal number of clusters is 4.**

```python
# Fit KMeans with selected number of clusters
kmeans = KMeans(n_clusters=4, init='k-means++', random_state=0)
kmeans.fit(df)

# Assign clusters to each customer
clusters = kmeans.predict(df)

# Add cluster labels to original DataFrame
df['kmean-cluster'] = clusters
```

**Display the mean values of the features for each cluster and investigate the peculiarities of each cluster.**

```python
df.groupby('kmean-cluster').mean()
```

**Specific more the range of age and income of 4 groups customer above**

```python
# Compute centroids
centroids = df.groupby('kmean-cluster').mean()

# Create scatter plot for Age vs Income with different colors for each cluster
plt.figure(figsize=(10, 7))

# Define color map and cluster names
colors = {0:'green', 1:'orange', 2:'blue', 3: 'purple'}
cluster_names = {0: 'Skilled Middle-age Single Male', 1: 'Unemployed Middle-age Non-single Female', 2: 'Highly Qualified Middle-age Single Male', 3: 'Skilled Middle-age Non-single Female'}
centroid_colors = {0:'green', 1:'orange', 2:'blue', 3: 'purple'}

# Create scatter plot
for cluster in df['kmean-cluster'].unique():
    plt.scatter(df[df['kmean-cluster']==cluster]['Age'], 
                df[df['kmean-cluster']==cluster]['Income'], 
                c=colors[cluster], 
                label=cluster_names[cluster])

# Add centroids to plot
for cluster in centroids.index:
    plt.scatter(centroids.loc[cluster, 'Age'], 
                centroids.loc[cluster, 'Income'], 
                c=centroid_colors[cluster], 
                marker='*', 
                s=600, 
                edgecolor='black', 
                linewidths=2)

plt.title('Age and Income Scatter Plot in KMeans method')
plt.xlabel('Age')
plt.ylabel('Income')
plt.legend()
plt.show()
```

**4.2 Hierarchical method**

**Import agglomerative clustering, dendrogram  and linkage**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from scipy.cluster.hierarchy import dendrogram, linkage
from sklearn.cluster import AgglomerativeClustering


# Run Hierarchical clustering
hie = AgglomerativeClustering(n_clusters=4, affinity='euclidean', linkage='ward')
clusters2 = hie.fit_predict(df)

# Add the cluster assignments to the original DataFrame
df['hierarchical-cluster'] = clusters2
```

**Perform hierarchical clustering and plot the dendrogram**

```python
from scipy.cluster.hierarchy import linkage, dendrogram
import matplotlib.pyplot as plt


# Set the color palette to 'Oranges'
sns.set_palette('Oranges')

# Perform hierarchical clustering and obtain the linkage matrix
linked = linkage(df, 'ward')

# Plot the dendrogram
plt.figure(figsize=(10, 7))
dendrogram(linked,
           orientation='top',
           distance_sort='descending',
           show_leaf_counts=True,
           color_threshold=0,
           above_threshold_color='gray')
plt.title('Hierarchical Cluster Dendrogram')
plt.xlabel('Sample index')
plt.ylabel('Euclidean distance')

# Show the plot
plt.show()
```

```python
# Scatterplot of Age and Income based on Hierarchical Clustering
x_axis = df["Age"]
y_axis = df["Income"]

plt.figure(figsize=(10, 8))
plt.title('Scatter Plot of Age and Income with Hierarchical Clusters')
sns.scatterplot(x_axis, y_axis, hue=df['hierarchical-cluster'], palette='Oranges')

plt.show()
```

```python
import matplotlib.pyplot as plt
from scipy.cluster import hierarchy


# Calculate the linkage matrix for hierarchical clustering
row_clusters = hierarchy.linkage(df.values, method='complete', metric='euclidean')

# Plot row dendrogram
fig = plt.figure(figsize=(12, 10), facecolor='white')
axd = fig.add_axes([0.09, 0.1, 0.2, 0.6])
row_dendr = hierarchy.dendrogram(row_clusters, orientation='left')
axd.set_xticks([])
axd.set_yticks([])
for spine in axd.spines.values():
    spine.set_visible(False)

# Reorder data with respect to clustering
#df_rowclust = df.iloc[row_dendr['leaves'][::-1]]

# Plot heatmap
#axm = fig.add_axes([0.23, 0.1, 0.6, 0.6])
#cax = axm.matshow(df_rowclust, interpolation='nearest', cmap='YlGnBu')
#fig.colorbar(cax)
#axm.set_xticklabels([''] + list(df_rowclust.columns), rotation=90)
#axm.set_yticklabels([''] + list(df_rowclust.index))

# Adjust spacing between subplots
#plt.subplots_adjust(left=0.3, top=0.9, right=0.9)

plt.show()
```
